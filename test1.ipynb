```python
import numpy as np
import pandas as pd
import talib
import datetime as datetime
import pandas_datareader as pdr
import pandas as pd
import os
pd.set_option('display.max_columns', None)
pd.set_option('display.max_rows', None)
pd.set_option('max_colwidth',100)
```

```python
data  = pd.read_csv("./train_lose_label.csv")
```

```python
time = list(data['public_date'])
y_label= [] 
for ind, name in  enumerate(data['TICKER']):
    date_now = str(time[ind])
    print(date_now)
    try:
        start = datetime.datetime( int(date_now[:4]),int(date_now[4:6]), int(date_now[6:]))
        now_price = pdr.DataReader(name ,'yahoo', start=start, end =start)['Close']
    except:
        try:
            start = start +datetime.timedelta(days=1)
            now_price = pdr.DataReader(name ,'yahoo', start=start, end =start)['Close']
        except:
            try:
                start = start +datetime.timedelta(days=1)
                now_price = pdr.DataReader(name ,'yahoo', start=start, end =start)['Close']
            except:
                print("error")
                y_label.append(-1)
                continue
#     print(start)
    after_start = start +datetime.timedelta(days= 90)
#     print(after_start)
    try:
        after_price = pdr.DataReader(name ,'yahoo', start=after_start, end =after_start)['Close']
    except:
        try:
            after_start = start +datetime.timedelta(days= 91)
            after_price = pdr.DataReader(name ,'yahoo', start=after_start, end =after_start)['Close']
        except:
            try:
                after_start = start +datetime.timedelta(days= 92)
                after_price = pdr.DataReader(name ,'yahoo', start=after_start, end =after_start)['Close']
            except:
                print("error")
                y_label.append(-1)
                continue
    print(now_price, after_price)
    now_price = float(now_price[0])
    after_price = float(after_price[0])
    if after_price >now_price *1.2:
         y_label.append(5)
    elif after_price >now_price *1.07:
        y_label.append(3)
    elif after_price >now_price *1.05:
        y_label.append(1)
    else:
        y_label.append(0)
```

```python
#x, y
x = [] 
y= []
for i in range(len(data)):
    if data['y_label'][i]== -1:
        continue
    else:
        x.append(list(np.array((data[i:i+1])))[0][:-1])
        if int(data['y_label'][i]) == 0:
            y.append(0)
        else:
            y.append(1)
#         y.append(int(data['y_label'][i]))
```

```python
# train_x = x[:int(8*(len(x))/10)]
# test_x = x[int(8*(len(x))/10):]
# train_y = y[:int(8*(len(x))/10)]
# test_y = y[int(8*(len(x))/10):]
from sklearn.model_selection import train_test_split
from xgboost import XGBClassifier
from sklearn.metrics import accuracy_score
```

```python
train_x, test_x, train_y, test_y = train_test_split(np.array(x),np.array(y), test_size=0.2, random_state=48)
```

```python
model = XGBClassifier()
```

```python
model.fit(train_x, train_y)
```

```python
y_pred = model.predict(test_x)
```

```python
accuracy = accuracy_score(test_y ,y_pred)
print(" ", accuracy*100,"%")
```
